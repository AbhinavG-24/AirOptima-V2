```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib

from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```
# ensure datetime
df['date'] = pd.to_datetime(df['date'])

# sort by time (VERY IMPORTANT)
df = df.sort_values('date').reset_index(drop=True)

# remove impossible pm10 values
df = df[df['pm10'] >= 0]

print("Data shape:", df.shape)
```

```
df['hour'] = df['date'].dt.hour
df['day_of_week'] = df['date'].dt.dayofweek
```

```
df['pm10_lag1'] = df['pm10'].shift(1)
df['pm10_lag2'] = df['pm10'].shift(2)
df['pm10_lag3'] = df['pm10'].shift(3)
df['pm10_lag6'] = df['pm10'].shift(6)
```

```
df['pm10_t+4'] = df['pm10'].shift(-4)
```

```
df_model = df.dropna().reset_index(drop=True)

print("After feature engineering:", df_model.shape)
```

```
FEATURES = [
    'pm10_lag1',
    'pm10_lag2',
    'pm10_lag3',
    'pm10_lag6',
    'hour',
    'day_of_week',
    'pm2_5'  # remove if too many NaNs
]

TARGET = 'pm10_t+4'

X = df_model[FEATURES]
y = df_model[TARGET]
```

```
train_size = int(len(df_model) * 0.8)

X_train = X.iloc[:train_size]
X_test = X.iloc[train_size:]

y_train = y.iloc[:train_size]
y_test = y.iloc[train_size:]

print("Train size:", X_train.shape)
print("Test size:", X_test.shape)
```

```
model = XGBRegressor(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.05,
    random_state=42
)

model.fit(X_train, y_train)
```

```
y_pred = model.predict(X_test)
```

```
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

```
plt.figure(figsize=(12,5))
plt.plot(y_test.values[:200], label="Actual", linewidth=2)
plt.plot(y_pred[:200], label="Predicted", linewidth=2)
plt.title("PM10 Actual vs Predicted (4-hour ahead)")
plt.legend()
plt.grid(True)
plt.show()
```

```
def get_pm10_risk(pm10):
    if pm10 > 430:
        return "Severe"
    elif pm10 > 350:
        return "Very Poor"
    elif pm10 > 250:
        return "Poor"
    elif pm10 > 150:
        return "Moderate"
    elif pm10 > 50:
        return "Satisfactory"
    else:
        return "Good"
```

```
df_model['risk_level'] = df_model['pm10'].apply(get_pm10_risk)

print(df_model[['pm10','risk_level']].head())
```

```
def get_alert_level(pred_pm10):
    if pred_pm10 > 250:
        return "HIGH RISK – Deploy Sprinkler"
    elif pred_pm10 > 150:
        return "MODERATE RISK – Monitor"
    else:
        return "LOW RISK – Normal"
```

```
df_model['pm_ratio'] = df_model['pm2_5'] / (df_model['pm10'] + 1e-6)
```

```
def identify_source(row):
    pm_ratio = row['pm_ratio']

    # 🚧 Dust / construction
    # detect unusually coarse events relative to your dataset
    if (row['pm10'] > 300) and (pm_ratio < 0.75):
        return "Road Dust / Construction"

    # 🚗 Vehicular emissions (dominant combustion)
    elif (pm_ratio > 0.78) and (row['no2'] > 60) and (row['co'] > 1500):
        return "Vehicular Emissions"

    # 🏭 Industrial / coal
    elif row['so2'] > 55:
        return "Industrial / Coal"

    # 🔥 Biomass burning
    elif (row['pm2_5'] > 160) and (row['co'] > 2200):
        return "Biomass/Waste Burning"

    else:
        return "Mixed/Uncertain"
```

```
df_model['likely_source'] = df_model.apply(identify_source, axis=1)

df_model[['pm10','pm2_5','pm_ratio','likely_source']].head()
```
